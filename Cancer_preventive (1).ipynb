```python
import os
import cv2
import numpy as np

def find_empty_masks(base_path, mask_suffix='_mask.png'):
    empty_masks = []
    folders = [f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))]

    for folder in folders:
        folder_path = os.path.join(base_path, folder)
        files = os.listdir(folder_path)

        image_files = [f for f in files if not f.endswith(mask_suffix)]

        for img_file in image_files:
            img_path = os.path.join(folder_path, img_file)
            mask_file = img_file.replace('.png', mask_suffix)
            mask_path = os.path.join(folder_path, mask_file)

            if not os.path.exists(mask_path):
                print(f"Mask not found for {img_file} in {folder}")
                continue

            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is None:
                print(f"Cannot read mask {mask_file} in {folder}")
                continue

            if np.count_nonzero(mask) == 0:
                empty_masks.append(mask_path)

    return empty_masks

base_path = './dataset/dataset/Dataset_BUSI_with_GT'
empty_masks_list = find_empty_masks(base_path)
print(f"Found {len(empty_masks_list)} empty masks across all folders.")
for mask_file in empty_masks_list:
    print(mask_file)
```

```python
import os
import cv2
import numpy as np

def summarize_empty_masks(base_path, mask_suffix='_mask.png'):
    folders = [f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))]
    summary = {}

    for folder in folders:
        folder_path = os.path.join(base_path, folder)
        files = [f for f in os.listdir(folder_path) if f.endswith('.png') and not f.endswith(mask_suffix)]
        empty_count = 0
        total_files = len(files)

        for img_file in files:
            mask_file = img_file.replace('.png', mask_suffix)
            mask_path = os.path.join(folder_path, mask_file)

            if not os.path.exists(mask_path):
                continue

            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
            if mask is None:
                continue

            if np.count_nonzero(mask) == 0:
                empty_count += 1

        summary[folder] = {'total_images': total_files, 'empty_masks': empty_count}

    return summary

base_dir = './dataset/dataset/Dataset_BUSI_with_GT'
summary_report = summarize_empty_masks(base_dir)

for folder, stats in summary_report.items():
    print(f"Folder: {folder}")
    print(f"Total images: {stats['total_images']}")
    print(f"Empty masks: {stats['empty_masks']}\n")
```

```python
import os
import cv2

def clean_masks(folder_path, mask_suffix='_mask.png'):
    files = [f for f in os.listdir(folder_path) if f.endswith('.png') and not f.endswith(mask_suffix)]

    for img_file in files:
        mask_file = img_file.replace('.png', mask_suffix)
        mask_path = os.path.join(folder_path, mask_file)

        if not os.path.exists(mask_path):
            print(f"Mask not found for image: {img_file} in {folder_path}")
            continue

        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        if mask is None:
            print(f"Cannot read mask: {mask_file} in {folder_path}")
            continue

        processed_mask = cv2.medianBlur(mask, 5)
        cv2.imwrite(mask_path, processed_mask)
        print(f"Cleaned mask: {mask_path}")

base_dir = './dataset/dataset/Dataset_BUSI_with_GT'

for folder_name in ['benign', 'malignant']:
    folder_path = os.path.join(base_dir, folder_name)
    clean_masks(folder_path)
```

```python
import os
import cv2
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
import numpy as np

class UltrasoundDataset(Dataset):
    def __init__(self, image_dir, image_list):
        self.image_dir = image_dir
        self.image_list = image_list

    def __len__(self):
        return len(self.image_list)

    def __getitem__(self, idx):
        img_name = self.image_list[idx]
        img_path = os.path.join(self.image_dir, img_name)
        mask_name = img_name.replace('.png', '_mask.png')
        mask_path = os.path.join(self.image_dir, mask_name)

        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

        image = cv2.resize(image, (256, 256))
        mask = cv2.resize(mask, (256, 256))

        image = image.astype(np.float32) / 255.0
        mask = (mask > 127).astype(np.float32)

        image = np.expand_dims(image, axis=0)
        mask = np.expand_dims(mask, axis=0)

        return torch.tensor(image, dtype=torch.float32), torch.tensor(mask, dtype=torch.float32)

base_dir = './dataset/dataset/Dataset_BUSI_with_GT/benign'

all_images = [f for f in os.listdir(base_dir) if f.endswith('.png') and not f.endswith('_mask.png')]

train_imgs, val_imgs = train_test_split(all_images, test_size=0.2, random_state=42)

train_dataset = UltrasoundDataset(base_dir, train_imgs)
val_dataset = UltrasoundDataset(base_dir, val_imgs)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        self.dconv_down1 = DoubleConv(1, 64)
        self.dconv_down2 = DoubleConv(64, 128)
        self.dconv_down3 = DoubleConv(128, 256)
        self.dconv_down4 = DoubleConv(256, 512)

        self.maxpool = nn.MaxPool2d(2)
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)

        self.dconv_up3 = DoubleConv(256 + 512, 256)
        self.dconv_up2 = DoubleConv(128 + 256, 128)
        self.dconv_up1 = DoubleConv(128 + 64, 64)

        self.conv_last = nn.Conv2d(64, 1, kernel_size=1)

    def forward(self, x):
        conv1 = self.dconv_down1(x)
        x = self.maxpool(conv1)

        conv2 = self.dconv_down2(x)
        x = self.maxpool(conv2)

        conv3 = self.dconv_down3(x)
        x = self.maxpool(conv3)

        x = self.dconv_down4(x)

        x = self.upsample(x)
        x = torch.cat([x, conv3], dim=1)
        x = self.dconv_up3(x)

        x = self.upsample(x)
        x = torch.cat([x, conv2], dim=1)
        x = self.dconv_up2(x)

        x = self.upsample(x)
        x = torch.cat([x, conv1], dim=1)
        x = self.dconv_up1(x)

        out = self.conv_last(x)
        return torch.sigmoid(out)
```

```python
import matplotlib.pyplot as plt
import cv2
import torch
import torch.nn as nn 
import torch.nn.functional as F 
import numpy as np
import os

class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.conv(x)

class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        self.dconv_down1 = DoubleConv(1, 64)
        self.dconv_down2 = DoubleConv(64, 128)
        self.dconv_down3 = DoubleConv(128, 256)
        self.dconv_down4 = DoubleConv(256, 512)

        self.maxpool = nn.MaxPool2d(2)
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)

        self.dconv_up3 = DoubleConv(256 + 512, 256)
        self.dconv_up2 = DoubleConv(128 + 256, 128)
        self.dconv_up1 = DoubleConv(128 + 64, 64)

        self.conv_last = nn.Conv2d(64, 1, kernel_size=1)

    def forward(self, x):
        conv1 = self.dconv_down1(x)
        x = self.maxpool(conv1)

        conv2 = self.dconv_down2(x)
        x = self.maxpool(conv2)

        conv3 = self.dconv_down3(x)
        x = self.maxpool(conv3)

        x = self.dconv_down4(x)

        x = self.upsample(x)
        x = torch.cat([x, conv3], dim=1)
        x = self.dconv_up3(x)

        x = self.upsample(x)
        x = torch.cat([x, conv2], dim=1)
        x = self.dconv_up2(x)

        x = self.upsample(x)
        x = torch.cat([x, conv1], dim=1)
        x = self.dconv_up1(x)

        out = self.conv_last(x)
        return torch.sigmoid(out)


def predict_and_visualize(model, image_path, device, mask_suffix='_mask.png'):
    """
    Predicts the mask for a given image and visualizes the original image and predicted mask.
    Optionally visualizes the ground truth mask if available.
    """
    model.eval()

    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Could not load image: {image_path}")
        return

    image = cv2.resize(image, (256, 256))
    image = image.astype(np.float32) / 255.0
    image = np.expand_dims(image, axis=0) 
    image_tensor = torch.tensor(image, dtype=torch.float32).unsqueeze(0).to(device) 

    with torch.no_grad():
        pred = model(image_tensor)
    predicted_mask = (pred.squeeze().cpu().numpy() > 0.5).astype(float)

    ground_truth_mask = None
    mask_candidates = [
        image_path.replace('.png', '_mask.png'),
        image_path.replace('.png', '_mask_1_mask.png'),
        image_path.replace('.png', '_mask_2_mask.png'),
    ]

    mask_path = None
    for candidate in mask_candidates:
        if os.path.exists(candidate):
            mask_path = candidate
            break

    if mask_path:
        ground_truth_mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        if ground_truth_mask is not None:
            ground_truth_mask = cv2.resize(ground_truth_mask, (256, 256))
            ground_truth_mask = (ground_truth_mask > 127).astype(float)


    num_plots = 2 if ground_truth_mask is None else 3
    plt.figure(figsize=(5 * num_plots, 5))

    plt.subplot(1, num_plots, 1)
    plt.imshow(image.squeeze(), cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    if ground_truth_mask is not None:
        plt.subplot(1, num_plots, 2)
        plt.imshow(ground_truth_mask, cmap='gray')
        plt.title('Ground Truth Mask')
        plt.axis('off')

    plt.subplot(1, num_plots, num_plots)
    plt.imshow(predicted_mask, cmap='gray')
    plt.title('Predicted Mask')
    plt.axis('off')

    plt.show()


test_image_path = './downloaded_images/downloaded_test_image.png'

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = UNet() 
model.to(device)


predict_and_visualize(model, test_image_path, device)
```
